```python
data_cumulative = covid_19_india.groupby('Date').sum()
data_cumulative.reset_index(inplace = True)

data_cumulative['Date'] = data_cumulative['Date'].apply(lambda data: '20' + '-'.join(data.split('/')[::-1]))

data_cumulative.sort_values('Date', inplace = True)

data_cumulative['Active'] = data_cumulative['Confirmed'] - (data_cumulative['Cured'] + data_cumulative['Deaths'])

data_cumulative = data_cumulative[['Date', 'Confirmed', 'Cured', 'Deaths', 'Active']]

data_cumulative.head()
```

```python
plt.figure(figsize=(18,8),dpi=100)
sns.lineplot(x ='Date', y ='Confirmed', data = data_cumulative,  label = 'Confirmed', color = 'blue', marker= 'o')
sns.lineplot(x = 'Date', y='Cured', data = data_cumulative, label = 'Cured', color = 'green')
sns.lineplot(x = 'Date', y='Deaths', data = data_cumulative, label = 'Deaths', color = 'black')
sns.lineplot(x = 'Date', y='Active', data = data_cumulative, label = 'Active', color = 'red')
plt.title('Cumulative number of cases (categorized by current health status) grouped by date')
plt.ylabel('Number of cases')
plt.legend(loc = 0)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```
